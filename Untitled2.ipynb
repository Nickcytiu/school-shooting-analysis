```python
def connect_to_endpoint(url, headers, params, next_token = None):
    params['next_token'] = next_token   #params object received from create_url function
    response = requests.request("GET", url, headers = headers, params = params)
    print("Endpoint Response Code: " + str(response.status_code))
    if response.status_code != 200:
        raise Exception(response.status_code, response.text)
    return response.json()
```

```python
def extract_twit_data(init):
    
    date = init["created_at"]
    text = init["text"]

    user = init["user"]
    name = user['name']
    screen_name = user["screen_name"]
    location = user["location"]
    verified = user["verified"]
    followers = user["followers_count"]
        
    return [date, text, name, screen_name, location, verified, followers]
```

```python
def parse_response(cols):
    
    for ix, re in enumerate(response["results"]):

        twit_dict[ix] = extract_twit_data(re)
        df_twit = pd.DataFrame.from_dict(twit_dict,
                                         orient='index',
                                         columns=cols)
        
        df_twit = df_twit.sort_values(by="Date")
    return df_twit
```
